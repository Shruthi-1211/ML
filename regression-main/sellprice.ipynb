```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('train.csv')
print("Dataset Preview:")
print(df.head())
```

```python
X = df[["GrLivArea", "BedroomAbvGr", "FullBath"]]
y = df["SalePrice"]
model = LinearRegression()
model.fit(X, y)

print("Multiple Regression model trained successfully.")
```

```python
print(f"\nCoefficients (GrLivArea, BedroomAbvGr, FullBath): {model.coef_}")
print(f"Intercept: {model.intercept_:.4f}")

# Visualize relationships
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 6))

sns.regplot(x='GrLivArea', y='SalePrice', data=df, ax=ax1, color='#e74c3c')
ax1.set_title('LiveArea vs Sale price')

sns.regplot(x='BedroomAbvGr', y='SalePrice', data=df, ax=ax2, color='#3498db')
ax2.set_title('Bedroom vs Saleprice')

sns.regplot(x='FullBath', y='SalePrice', data=df, ax=ax3, color='#3498db')
ax3.set_title('Fullbath vs Saleprice')

plt.show()
```

```python
def predict_price(GrLivArea, BedroomAbvGr, FullBath):
    prediction = model.predict([[GrLivArea, BedroomAbvGr, FullBath]])[0]
    print(f"\n Predicted Selling Price:")
    print(f"Live Area: {GrLivArea}, Bedroom: {BedroomAbvGr}, FullBath{FullBath}")
    print(f"Estimated Selling Price: {prediction:.2f}")
```

```python
try:
    u_GrLivArea = float(input("Enter Living Area (sq ft): "))
    u_BedroomAbvGr = int(input("Enter Number of Bedrooms: "))
    u_FullBath = int(input("Enter Number of Full Bathrooms: "))

    predict_price(u_GrLivArea, u_BedroomAbvGr, u_FullBath)

except ValueError:
    print("Please enter valid numeric values.")
```
